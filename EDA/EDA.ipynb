```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load Data
file_path = 'ChemBL2599_combined_data.csv.gz' # Ensure this filename is correct
df = pd.read_csv(file_path)
df_clean = df.dropna(subset=['pChEMBL Value']).copy()

# 2. Prepare Data for Correlation
# Select relevant numerical columns
cols_to_corr = [
    'pChEMBL Value',
    'Molecular Weight',
    'AlogP',
    '#RO5 Violations',
    'Ligand Efficiency BEI',
    'Ligand Efficiency LE'
]

# Ensure columns are numeric (handling any potential non-numeric errors)
for col in cols_to_corr:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

# 3. Correlation Heatmap
plt.figure(figsize=(10, 8))
corr_matrix = df_clean[cols_to_corr].corr()
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=.5,
    center=0
)
plt.title('Correlation Heatmap: Potency vs Properties')
plt.tight_layout()
plt.show()

# 4. Box Plot: Potency vs RO5 Violations
# This helps visualize if "rule-breaking" compounds are more or less potent
plt.figure(figsize=(10, 6))
sns.boxplot(
    x='#RO5 Violations',
    y='pChEMBL Value',
    data=df_clean,
    palette='Set2'
)
plt.title('Potency Distribution (pChEMBL) grouped by #RO5 Violations')
plt.xlabel('Number of RO5 Violations')
plt.ylabel('pChEMBL Value')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 5. Box Plot: Molecular Weight Distribution
plt.figure(figsize=(10, 4))
sns.boxplot(x=df_clean['Molecular Weight'], color='skyblue')
plt.title('Distribution of Molecular Weight')
plt.xlabel('Molecular Weight (Daltons)')
plt.show()
```
